```python
df = pd.read_csv('weather.csv', encoding='utf8', on_bad_lines='skip', sep=';')
```

```python
df.head(10)
```

```python
df['month'] = pd.DatetimeIndex(df['Местное время в Харькове (аэропорт)']).month
df['day'] = pd.DatetimeIndex(df['Местное время в Харькове (аэропорт)']).day
df['week'] = pd.to_datetime(df['Местное время в Харькове (аэропорт)']).dt.isocalendar().week
```

```python
# day with the highest temp
df.loc[df.groupby(["day"])["T"].idxmax()] 
```

```python
# day with the lowest temp
df.loc[df.groupby(["day"])["T"].idxmin()] 
```

```python
df_month = df.groupby('month').mean()
```

```python
# month with the highest average wind speed
df_month['Ff'].idxmax()
```

```python
# month with the lowest average temperature
df_month['T'].idxmin()
```

```python
# month with the highest average temperature
df_month['T'].idxmax()
```

```python
# week with the highest amount of rain 
df['RRR'] = pd.to_numeric(df['RRR'], errors='coerce')
df.sort_values(['RRR'],ascending=False, inplace=True)
df
```
